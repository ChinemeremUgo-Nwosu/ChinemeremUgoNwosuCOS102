```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox

# Load the Excel file
df = pd.read_excel(r"GIG-logistics.xlsx")

# Normalize data (important for matching)
df['Name'] = df['Name'].str.lower().str.strip()
df['Department'] = df['Department'].str.lower().str.strip()

# Get unique departments
departments = sorted(df['Department'].unique())

# Function to search
def search():
    name = entry_name.get().lower().strip()
    dept = selected_dept.get()

    # Filter department
    dept_df = df[df['Department'] == dept]
    match = dept_df[dept_df['Name'] == name]

    if not match.empty:
        # Get other members
        others = dept_df[dept_df['Name'] != name]['Name'].tolist()

        message = f"{name.title()} is in {dept.title()} department.\n\nOther members:\n"
        message += "\n".join([person.title() for person in others])

        messagebox.showinfo("Success", message)
    else:
        messagebox.showwarning("Warning", f"{name.title()} not found in {dept.title()} department.")

# GUI setup
root = tk.Tk()
root.title("GIG Employee Search")
root.geometry("320x320")

tk.Label(root, text="Enter Name").pack()
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Select Department").pack()

# Dropdown variable
selected_dept = tk.StringVar()
selected_dept.set(departments[0])  # default value

# Dropdown menu
dept_menu = tk.OptionMenu(root, selected_dept, *departments)
dept_menu.pack()

tk.Button(root, text="Search", command=search).pack()

root.mainloop()
```
