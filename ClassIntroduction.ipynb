```python
class SST:
    prog1 = 'Computer Science' # variable for class

    def __init__(self, name):
        self.name = name

    def speak(self):
        print('\n My name is', self.name, 'and I study', self.prog1)

stud1 = SST(input('Enter the name for the first student: '))
stud2 = input('Enter the name for the second student: ')

person2 = SST(stud2)
stud1.speak()
person2.speak()
```

```python
class Addition:
    def __init__(self, f, s):
        self.first = f
        self.second = s
        self.answer = None

    def calculate(self):
        self.answer = self.first + self.second

    def display(self):
        print('\nFirst:', self.first)
        print('Second:', self.second)
        print('\nSum:', self.answer)


num1 = int(input('Enter the first number: '))
num2 = int(input('Enter the second number: '))

obj = Addition(num1, num2)
obj.calculate()
obj.display()
```
